```python
# Import libraries
import pandas as pd
```

```python
# Load the dataset, accepted loans only
file_path = "../data/accepted_2007_to_2018Q4.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# Check the shape of the DataFrame
df.shape
```

```python
# Examine the target variable

# df['loan_status'].unique()
print("\nLoan Status Counts:")
print(df['loan_status'].value_counts())
```

```python
# Convert issue date to datetime
df['issue_d'] = pd.to_datetime(df['issue_d'], errors='coerce')

# Count loans issued in 2017
df_2017 = df[df['issue_d'].dt.year == 2017]
print("\nLoans issued in 2017:", df_2017.shape[0])
df_2017.shape
```

```python
# write 2017 loans accepted to a new csv
df_2017.to_csv("../data/accepted_2017.csv", index=False)
```

```python
# reduce even further, down to loans with valid loan_status
valid_statuses = [
    "Fully Paid",
    "Charged Off",
    "Late (31-120 days)",
    "Late (16-30 days)",
    "Default"
]
df_2017_filtered = df_2017[df_2017['loan_status'].isin(valid_statuses)]
```

```python
# examine the target variable again
df_2017_filtered['loan_status'].value_counts()
```

```python
# emanine the shape of the filtered 2017 DataFrame
df_2017_filtered.shape
```

```python
# write this to a new csv
df_2017_filtered.to_csv("../data/accepted_2017_valid_targets.csv", index=False)
```

```python
# look for features with lots of missing values, first redefine dataframe
df = pd.read_csv("../data/accepted_2017_valid_targets.csv")
missing_percent = df.isnull().mean().sort_values(ascending=False) * 100
missing_percent
```

```python
# view features with more than 50% missing values
missing_percent[missing_percent > 50]
```

```python
# drop features with more than 50% missing values
threshold = 0.50   
cols_to_drop = missing_percent[missing_percent > threshold * 100].index.tolist()

df_reduced = df.drop(columns=cols_to_drop)
```

```python
df_reduced.shape
```

```python
# complete this notebook and exploration by writing reduced dataframe to csv
df_reduced.to_csv("../data/accepted_2017_reduced.csv", index=False)
```
