```python
# import necessary libraries
import pandas as pd
import numpy as np
import re
```

```python
# load accepted_2017_eda.csv dataset as dataframe
df = pd.read_csv("../data/accepted_2017_eda.csv")
```

```python
# look for fields in need of transformation with descriptive statistics
df.describe(include='all')
```

```python
# remove target_binary and loan_status_binary columns from eda set
df = df.drop(columns=["target_binary", "loan_status_binary"])
df.shape, df.columns
```

```python
# look for transformations again
df.describe(include='all')
```

```python
# convert term from text to integer
df["term"] = df["term"].str.extract(r"(\d+)").astype(int)
```

```python
# look for transofmrations again
df.describe(include='all')
```

```python
# drop grade column
df = df.drop(columns=["grade"])
df.shape, df.columns
```

```python
# unique values in sub_grade column
df["sub_grade"].unique()
```

```python
# Get unique values (optional, for inspection)
unique_subgrades = sorted(df["sub_grade"].dropna().unique())
print(unique_subgrades)

# Create mapping for letter → base number
letter_map = {'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7}

# Convert subgrade to an ordinal integer 1–35
df["sub_grade_num"] = (
    df["sub_grade"].str[0].map(letter_map) * 5
    + df["sub_grade"].str[1].astype(int)
    - 5     # adjust so A1 starts at 1 instead of 6
)

# Check the result
df[["sub_grade", "sub_grade_num"]].head(20)
```

```python
# look for transofmrations again
df.describe(include='all')
```

```python
# drop sub_grade column
df = df.drop(columns=["sub_grade"])
df.shape, df.columns
```

```python
# look for transformations again
df.describe(include='all')
```

```python
# unique values in emp_length column
df["emp_length"].unique()
```

```python
# transform emp_length to integer
def parse_emp_length(val):
    if pd.isna(val):
        return np.nan
    val = val.strip()
    if val == '10+ years':
        return 10
    if val == '< 1 year':
        return 0
    # catch '1 year', '2 years', ..., '9 years'
    m = re.search(r'(\d+)', val)
    if m:
        return int(m.group(1))
    return np.nan

df["emp_length_years"] = df["emp_length"].apply(parse_emp_length)
```

```python
# look for transformations again
df.describe(include='all')
```

```python
# drop emp_length column
df = df.drop(columns=["emp_length"])
df.shape, df.columns
```

```python
# look for transformations again
df.describe(include='all')
```

```python
# one hot encode home_ownership column
home_ownership_dummies = pd.get_dummies(df["home_ownership"], prefix="home")
df = pd.concat([df, home_ownership_dummies], axis=1)
```

```python
# determine if target is missing any values
df["loan_status_binary"].isnull().sum()
```

```python
# handle true/false cols
# Find all boolean columns
bool_cols = df.select_dtypes(include='bool').columns

# Convert them to integers (True → 1, False → 0)
df[bool_cols] = df[bool_cols].astype(int)
```

```python
# look at description again
df.describe(include='all')
```

```python
# find features with missing values
missing_cols = df.columns[df.isna().any()]
df[missing_cols].isna().sum().sort_values(ascending=False)
```

```python
# assume when il_util is missing it means 0
df["il_util"] = df["il_util"].fillna(0)
# assume when mths_since_recent_inq is missing it means no recent inquiries
df["mths_since_recent_inq"] = df["mths_since_recent_inq"].fillna(0)
# emp_length_years missing values can be filled with -1 as the customer didn't enter the value, it is it's own category
df["emp_length_years"] = df["emp_length_years"].fillna(-1)
# assume num_tl_120dpd_2m missing values means 0
df["num_tl_120dpd_2m"] = df["num_tl_120dpd_2m"].fillna(0)
# assume mths_since_rcnt_il means they never had an installment loan
df["mths_since_rcnt_il"] = df["mths_since_rcnt_il"].fillna(0)
#  assume mo_sin_old_il_acct means they never had an installment loan
df["mo_sin_old_il_acct"] = df["mo_sin_old_il_acct"].fillna(0)
#  assume the same for bc_util, percent_bc_gt_75, bc_open_to_buy, mths_since_recent_bc
df["bc_util"] = df["bc_util"].fillna(0)
df["percent_bc_gt_75"] = df["percent_bc_gt_75"].fillna(0)
df["bc_open_to_buy"] = df["bc_open_to_buy"].fillna(0)
df["mths_since_recent_bc"] = df["mths_since_recent_bc"].fillna(0)
```

```python
# find features with missing values
missing_cols = df.columns[df.isna().any()]
df[missing_cols].isna().sum().sort_values(ascending=False)
```

```python
# fill missing dti with median
df["dti"] = df["dti"].fillna(df["dti"].median())
# fill revol_util with 0s
df["revol_util"] = df["revol_util"].fillna(0)
# fill all_util with 0s
df["all_util"] = df["all_util"].fillna(0)
# avg_cur_bal with 0s
df["avg_cur_bal"] = df["avg_cur_bal"].fillna(0)
```

```python
# find features with missing values
missing_cols = df.columns[df.isna().any()]
df[missing_cols].isna().sum().sort_values(ascending=False)
```
